```python
#import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
#plot a sample
plt.plot([1,2,3,4], [1,4,8,10])
plt.show()
```

```python
#add title
plt.plot([1,2,3,4], [1,4,8,10])
plt.title("sample plot")
plt.xlabel('X label')
plt.ylabel('Y label')
plt.show()
```

```python
#config size
fig = plt.figure(figsize=(15,5))
plt.plot([1,2,3,4], [1,4,8,10])
plt.title("sample plot")
plt.xlabel('X label')
plt.ylabel('Y label')
plt.show()
```

```python
#config type of graph
fig = plt.figure(figsize=(15,5))
plt.plot([1,2,3,4], [1,4,8,10], "go")
plt.title("sample plot")
plt.xlabel('X label')
plt.ylabel('Y label')
plt.show()
```

```python
#draw multi data source
x = np.arange(1,5)
y = x ** 3
plt.plot([1,2,3,4], [1,4,8,10], "go", x, y, "r^")
plt.title("sample plot")
plt.xlabel('X label')
plt.ylabel('Y label')
plt.show()
```

```python
#multi subplot, one row
plt.subplot(1,2,1)
plt.plot([1,2,3,4], [1,4,8,10], "go")
plt.title("sample subplot 1")

plt.subplot(1,2,2)
plt.plot(x, y, "r^")
plt.title("sample subplot 2")

plt.suptitle("Main title")
plt.show()
```

```python
#multi subplot, two row
plt.subplot(2,1,1)
plt.plot([1,2,3,4], [1,4,8,10], "go")
plt.title("sample subplot 1")

plt.subplot(2,1,2)
plt.plot(x, y, "r^")
plt.title("sample subplot 2")

plt.suptitle("Main title")
plt.show()
```

```python
#subplot with built in function
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(6,6))
ax[0,1].plot([1,2,3,4], [1,4,8,10], "go")
ax[1,0].plot(x, y, "r^")
ax[0,1].set_title("sample subplot 1")
ax[1,0].set_title("sample subplot 2")
plt.show()
```
